```python
import re
import io
#import os
#import json
import string
import numpy as np
from gensim.models import Word2Vec
```

### Import necessary files here

### Load the data file

```python
def load_corpus(filename):
    print(filename+": \n")
    with io.open(filename, 'r', encoding="utf-8") as f:
        corpus = f.read()
    return corpus
```

```python
def clean_data(data):
    dirt = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ.-,?:1234567890(){}[]‘’১২৩৪৫৬৭৮৯০"
    data = [char for char in data if char not in dirt]
    data = ''.join(data)
    return data
```

```python
def tokenize(corpus):
    corpus = corpus.replace('\n',' ')
    corpus = corpus.split("।")
    corpus = [sentence.strip() for sentence in corpus]
    data = [sentence.split() for sentence in corpus]
    return data
```

```python
def remove_stopword(data):
    return data
```

```python
def print_data(data):
    print(type(data))
    print(len(data))
    print(data[:1000])
```

```python
corpus = load_corpus('newfile.txt')
corpus = clean_data(corpus)
corpus = tokenize(corpus)
print_data(corpus)
```

```python
book = load_corpus('book1.txt')
book = clean_data(book)
book = tokenize(book)
print_data(book)
```

```python
corpus.extend(book)
print(len(corpus))
```

```python
model_cbow = Word2Vec(corpus, size=200, window=5, min_count=5, sg=0)
```

```python
model_sg = Word2Vec(corpus, size=200, window=5, min_count=5, sg=1)
```

```python
val = model_cbow.wv.most_similar('বাংলাদেশ', topn=5)
print(val)
```

```python
val = model_sg.wv.most_similar('বাংলাদেশ', topn=5)
print(val)
```

```python
def data_preprocesses():
    data = []
    letter = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ-,?:1234567890(){}[]‘’১২৩৪৫৬৭৮৯০."
    with io.open('stop_word.txt', "r", encoding="utf-8") as file:
        filedata = file.read()

    data = filedata.split("\n")
    with io.open('newfile.txt', "r", encoding="utf-8") as file:
        filedata = file.read()
        # replace end of line with \n
        filedata = filedata.replace("। ", "\n")

    # remove all number
    filedata = ''.join(i for i in filedata if not i.isdigit())

    # remove bracket
    filedata = filedata.replace("(", "")
    filedata = filedata.replace(")", "")
    for word in data:
        filedata = filedata.replace(" " + word + " ", " ")

    filedata = re.sub(r'\n+', '\n', filedata).strip()

    filedata = [char for char in filedata if char not in letter]
    filedata = ''.join(filedata)

    # Write the file out again
    with io.open('newfile2.txt', "w", encoding="utf-8") as file:
        file.write(filedata)


data_preprocesses()
```
